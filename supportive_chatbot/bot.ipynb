```python
import nltk
import numpy as np
import random
import bs4 as bs
import urllib.request
import re
```

```python
link = urllib.request.urlopen('https://en.wikipedia.org/wiki/LeBron_James')
link = link.read()
data = bs.BeautifulSoup(link, 'lxml')
```

```python
data
```

```python
data_paragraphs = data.find_all('p')
data_text = ""
for para in data_paragraphs:
    data_text += para.text
```

```python
data_text
```

```python
data_text = data_text.lower()
data_text = re.sub(r'\[[0-9]*\]', ' ', data_text)
data_text = re.sub(r'\s+', ' ', data_text)
```

```python
data_text
```

```python
sen = nltk.sent_tokenize(data_text)
words = nltk.word_tokenize(data_text)
```
